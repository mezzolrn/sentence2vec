```python
r = open('../../data/total/vec2400_train.txt','r', encoding='utf-8').readlines()[:]
X_train = []
y_train = []

for i in r:
    y_train.append(i[0])
    tmp_list = i[2:].split(',')
    for index in range(len(tmp_list)):
        tmp_list[index] = float(tmp_list[index])
    X_train.append(tmp_list)

    
clf = svm.SVC(C=1.0, kernel='rbf')
clf.fit(X_train, y_train)

r = open('../../data/total/vec2400_test.txt','r', encoding='utf-8').readlines()[:]
X_test = []
y_test = []

for i in r:
    y_test.append(i[0])
    tmp_list = i[2:].split(',')
    for index in range(len(tmp_list)):
        tmp_list[index] = float(tmp_list[index])
    X_test.append(tmp_list)

r = clf.predict(X_test)
print(classification_report(y_test, r, digits=6))
```
